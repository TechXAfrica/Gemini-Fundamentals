### Import the necessary packages.

```python
import pathlib
import textwrap

import google.generativeai as genai

from IPython.display import display
from IPython.display import Markdown

# Used to securely store your API key
from google.colab import userdata
```

```python
#define a function to show responses generated by model
def to_markdown(text):
  text = text.replace('', '  *')
  return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
```

```python
#retrieve the key
GOOGLE_API_KEY=userdata.get('gdsc_germini')

genai.configure(api_key=GOOGLE_API_KEY)
```

### Text Generation

```python
model = genai.GenerativeModel(model_name = "gemini-pro")
model
```

```python
prompt = [
    "What advice would you give to other women who are just starting their journey in tech?",
]

response = model.generate_content(prompt)
```

```python
to_markdown(response.text)
```

### Using Gemini API with LangChains

```python
from langchain_google_genai import ChatGoogleGenerativeAI
```

```python
llm = ChatGoogleGenerativeAI(model="gemini-pro", google_api_key=GOOGLE_API_KEY)
```

```python
result = llm.invoke("What advice would you give to other women who are just starting their journey in tech?")
```

```python
to_markdown(result.content)
```

And that's it! You've now integrated Gemini Pro with LangChain!

```python
from langchain.prompts import PromptTemplate
from langchain.chains import LLMChain
```

```python
tweet_prompt = PromptTemplate.from_template("You are a content creator. Write me a tweet about {topic}.")

tweet_chain = LLMChain(llm=llm, prompt=tweet_prompt, verbose=True)

if __name__=="__main__":
    topic = "why Women Tech Makers is such a cool community"
    resp = tweet_chain.run(topic=topic)
    print(resp)
```

### Image Recognition

In this section, we will use the image from [WTM Nairobi](https://twitter.com/DeniseAllela/status/1720689892345168218/photo/2)

```python
from PIL import Image

img_path = "/content/Images/wtm_breakfast_2023.PNG"

# Open the image
img = Image.open(img_path)

# Display the image (optional)
display(img)
```

```python
model = genai.GenerativeModel('gemini-pro-vision')
response = model.generate_content(img)

to_markdown(response.text)
```

```python
response = model.generate_content(
    [
        "You are a blogger. Write a short, engaging blog post based on this picture. It should include a description of the objects in the photo and talk about what organizations can take to promote diversity and inclusion in the tech workplace.",
        img
    ],
    stream=True
)
response.resolve()
```

```python
to_markdown(response.text)
```

### Gemini Pro Vision

```python
from langchain_core.messages import HumanMessage
from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(model="gemini-pro-vision")
# example
message = HumanMessage(
    content=[
        {
            "type": "text",
            "text": "What's in this image?",
        },
        {"type": "image_url", "image_url": "https://picsum.photos/seed/picsum/200/300"},
    ]
)
llm.invoke([message])
```

```python
# prompt: How to show an image from a given url within markdown
import requests
import io
from PIL import Image

def show_image(url):
  response = requests.get(url)
  image = Image.open(io.BytesIO(response.content))
  display(image)

show_image("https://tinyurl.com/xpb6vjdc")
```
